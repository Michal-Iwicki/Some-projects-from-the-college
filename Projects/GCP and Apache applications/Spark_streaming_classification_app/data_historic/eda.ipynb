```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", context="talk")

# Display settings
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)
```

```python
df = pd.read_csv("snapshot_2024-09-01_to_2024-09-08.csv")
# Convert timestamps
df["flight_time"] = pd.to_datetime(df["flight_time"])
df["weather_time"] = pd.to_datetime(df["weather_time"])
df["datetime"] = pd.to_datetime(df["datetime"])

df.head()
```

```python
print("=== INFO ===")
df.info()

print("\n=== MISSING VALUES ===")
display(df.isna().sum())

print("\n=== SUMMARY STATISTICS ===")
display(df.describe(include="all"))
```

## Distribution of Delays

```python
plt.figure(figsize=(10,6))
sns.histplot(df["delay_minutes"], bins=40, kde=True, color="steelblue")
plt.title("Distribution of Flight Delays (minutes)")
plt.xlabel("Delay (minutes)")
plt.ylabel("Count")
plt.show()
```

## Delays by Airline

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="Carrier", y="delay_minutes")
plt.title("Flight Delays by Carrier")
plt.xlabel("Carrier")
plt.ylabel("Delay (minutes)")
# plt.yscale("log")
plt.show()
```

## Correlation Between Weather Variables and Delay

```python
weather_cols = ["temp", "humidity", "windgust", "windspeed", "pressure", "cloudcover", "visibility"]

corr = df[weather_cols + ["delay_minutes"]].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Weather Variables and Delay")
plt.show()
```

## Temperature vs Delay

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="temp", y="delay_minutes", hue="Carrier")
plt.title("Temperature vs Delay")
plt.xlabel("Temperature (C)")
plt.ylabel("Delay (minutes)")
plt.show()
```

## Wind Speed vs Delay

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="windspeed", y="delay_minutes", hue="Carrier")
plt.title("Wind Speed vs Delay")
plt.xlabel("Wind Speed (km/h)")
plt.ylabel("Delay (minutes)")
plt.show()
```

## Seismic Activity & Flight Delays


```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x="quake_count_6h",
    y="delay_minutes",
    hue="Carrier",
    size="max_mag_6h",
    sizes=(20,200)
)
plt.title("Seismic Activity (6h window) vs Flight Delay")
plt.xlabel("Earthquake Count (6h)")
plt.ylabel("Delay (minutes)")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left", borderaxespad=0)
plt.show()
```

##  TimeSeries Analysis

```python
plt.figure(figsize=(14,6))
sns.lineplot(data=df.sort_values("flight_time"), x="flight_time", y="delay_minutes")
plt.title("Flight Delays Over Time")
plt.xlabel("Flight Time")
plt.ylabel("Delay (minutes)")
plt.show()
```

```python
plt.figure(figsize=(14,6))

sns.scatterplot(
    data=df.sort_values("flight_time"),
    x="flight_time",
    y="delay_minutes",
    s=25,              # point size
    edgecolor="none",  # removes borders
    alpha=0.7          # optional: makes plot cleaner
)

plt.title("Flight Delays Over Time")
plt.xlabel("Flight Time")
plt.ylabel("Delay (minutes)")
plt.tight_layout()
plt.show()
```

## Categorical Distributions

```python
plt.figure(figsize=(10,6))
df["Carrier"].value_counts().plot(kind="bar", color="slateblue")
plt.title("Number of Flights per Carrier")
plt.xlabel("Carrier")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,6))
df["Dest"].value_counts().head(20).plot(kind="bar", color="teal")
plt.title("Top Destinations")
plt.xlabel("Destination")
plt.ylabel("Count")
plt.show()
```

## Weather Condition Frequencies

```python
plt.figure(figsize=(12,6))
df["conditions"].value_counts().plot(kind="bar", color="orange")
plt.title("Weather Conditions Frequency")
plt.xlabel("Condition")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(numeric_only=True), cmap="viridis", annot=False)
plt.title("Full Numeric Correlation Heatmap")
plt.show()
```
